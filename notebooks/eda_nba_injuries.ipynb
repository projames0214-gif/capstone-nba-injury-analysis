```python
import pandas as pd

# Load datasets
injury_df = pd.read_csv('../data/raw/NBA Player Injury Stats(1951 - 2023).csv')
stats_df = pd.read_csv('../data/raw/nba_player_stats_injuries.csv')

print("Injury dataset shape:", injury_df.shape)
print("Stats dataset shape:", stats_df.shape)
```

```python
# Preview injury dataset
print("=== INJURY DATASET ===")
print(injury_df.head())
print("\nColumns:", injury_df.columns.tolist())
print("\nData types:\n", injury_df.dtypes)
print("\nMissing values:\n", injury_df.isnull().sum())
```

```python
# Preview stats dataset
print("=== STATS DATASET ===")
print(stats_df.head())
print("\nColumns:", stats_df.columns.tolist())
print("\nData types:\n", stats_df.dtypes)
print("\nMissing values:\n", stats_df.isnull().sum())
```

```python
# Basic stats on key columns
print("=== INJURY DATASET - Notes sample ===")
print(injury_df['Notes'].value_counts().head(10))

print("\n=== STATS DATASET - Injury types ===")
print(stats_df['INJURED_TYPE'].value_counts().head(10))

print("\n=== STATS DATASET - Seasons covered ===")
print(stats_df['SEASON'].unique())

print("\n=== STATS DATASET - Average minutes per game ===")
print(stats_df['MIN'].describe())
```
